```python
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# Windows   
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc("font", family=font)

#    
plt.rcParams["axes.unicode_minus"] = False


#  
PROJECT_ROOT = Path.cwd().parents[0]  # notebooks 
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from financial_physics.metrics.msd import rolling_alpha

#  
lp = pd.read_parquet(PROJECT_ROOT / "data" / "processed" / "log_prices.parquet")

#   ()
series = lp["KRW=X"].dropna()
x = series.values
dates = series.index

# rolling alpha 
centers, alphas = rolling_alpha(
    x,
    window=252,      # 1
    step=5,          #  
    fit_range=(5, 120),
)

# centers  
center_dates = dates[centers]

# 
plt.figure(figsize=(12, 5))
plt.plot(center_dates, alphas, color='black', label="Rolling ")
plt.axhline(1.0, linestyle="--", color="gray", label=" = 1 ( )")

#    
plt.axvspan(pd.Timestamp("2008-09-01"), pd.Timestamp("2009-06-01"),
            color="red", alpha=0.1, label="2008 ")
plt.axvspan(pd.Timestamp("2020-02-01"), pd.Timestamp("2020-06-01"),
            color="blue", alpha=0.1, label="2020 ")
plt.axvspan(pd.Timestamp("2022-03-01"), pd.Timestamp("2022-12-01"),
            color="green", alpha=0.1, label="2022  ")

plt.title("USD/KRW Rolling MSD Exponent ()")
plt.xlabel("Date")
plt.ylabel("")
plt.legend()
plt.tight_layout()
plt.show()
```
